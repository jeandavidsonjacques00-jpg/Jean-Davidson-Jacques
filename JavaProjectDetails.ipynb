```python
%%writefile Product.java
public class Product {
    private String productName;
    private double price;
    private String category;
    private int quantityInStock;

    public Product(String productName, double price, String category, int quantityInStock) {
        this.productName = productName;
        this.price = price;
        this.category = category;
        this.quantityInStock = quantityInStock;
    }

    public String getProductName() { return productName; }
    public double getPrice() { return price; }
    public String getCategory() { return category; }
    public int getQuantityInStock() { return quantityInStock; }

    public void setPrice(double price) { this.price = price; }
    public void setQuantityInStock(int quantityInStock) { this.quantityInStock = quantityInStock; }

    public void displayProductDetails() {
        System.out.println("Product: " + productName +
                           ", Category: " + category +
                           ", Price: $" + price +
                           ", In stock: " + quantityInStock);
    }

    public boolean checkStock(int requestedQty) {
        return quantityInStock >= requestedQty;
    }

    public void updateStock(int quantity) {
        this.quantityInStock -= quantity;
    }
}

class Electronics extends Product {
    public Electronics(String name, double price, int quantityInStock) {
        super(name, price, "Electronics", quantityInStock);
    }
}

class Clothing extends Product {
    public Clothing(String name, double price, int quantityInStock) {
        super(name, price, "Clothing", quantityInStock);
    }
}
```

```python
%%writefile LoginInformation.java
public class LoginInformation {
    private String userName;
    private String password;
    private String debitCardNumber;
    private String debitCardZipCode;

    public LoginInformation(String userName, String password,
                            String debitCardNumber, String debitCardZipCode) {
        this.userName = userName;
        this.password = password;
        this.debitCardNumber = debitCardNumber;
        this.debitCardZipCode = debitCardZipCode;
    }

    public boolean validateLogin(String inputUser, String inputPass) {
        return this.userName.equals(inputUser) && this.password.equals(inputPass);
    }

    public String getUserName() { return userName; }
    public String getDebitCardNumber() { return debitCardNumber; }
    public String getDebitCardZipCode() { return debitCardZipCode; }
}
```

```python
%%writefile Cart.java
import java.util.ArrayList;
import java.util.List;

public class Cart {

    private static class CartItem {
        Product product;
        int quantity;

        CartItem(Product product, int quantity) {
            this.product = product;
            this.quantity = quantity;
        }
    }

    private List<CartItem> items = new ArrayList<>();

    public void addProduct(Product product, int quantity) {
        if (product.checkStock(quantity)) {
            items.add(new CartItem(product, quantity));
        } else {
            System.out.println("Not enough stock for " + product.getProductName());
        }
    }

    public void removeProduct(String productName) {
        items.removeIf(item -> item.product.getProductName().equalsIgnoreCase(productName));
    }

    public void displayCart() {
        if (items.isEmpty()) {
            System.out.println("Cart is empty.");
            return;
        }
        for (CartItem item : items) {
            System.out.println(item.product.getProductName() +
                               " x " + item.quantity +
                               " @ $" + item.product.getPrice());
        }
    }

    public void clearCart() {
        items.clear();
    }

    public List<CartItem> getItems() {
        return items;
    }

    public int getTotalProductCount() {
        int total = 0;
        for (CartItem item : items) {
            total += item.quantity;
        }
        return total;
    }
}
```

```python
%%writefile InsufficientFundsException.java
public class InsufficientFundsException extends Exception {
    public InsufficientFundsException(String message) {
        super(message);
    }
}

public class InvalidZipCodeException extends Exception {
    public InvalidZipCodeException(String message) {
        super(message);
    }
}

public class ProductNotFoundException extends Exception {
    public ProductNotFoundException(String message) {
        super(message);
    }
}
```

```python
%%writefile Tax.java
public class Tax {
    private static final double TAX_RATE = 0.07; // 7%

    public double calculateTax(double subtotal) {
        return subtotal * TAX_RATE;
    }

    public double totalWithTax(double subtotal) {
        return subtotal + calculateTax(subtotal);
    }
}
```

```python
%%writefile Payment.java
public class Payment {
    private double totalAmount;
    private boolean paymentStatus;

    public Payment(double totalAmount) {
        this.totalAmount = totalAmount;
        this.paymentStatus = false;
    }

    public boolean isPaymentSuccessful() {
        return paymentStatus;
    }

    public void acceptPayment(LoginInformation loginInfo, String addressZip, double availableBalance)
            throws InsufficientFundsException, InvalidZipCodeException {

        DebitCardValidator validator = new DebitCardValidator(loginInfo, addressZip, availableBalance);

        if (!validator.validateZipCode()) {
            throw new InvalidZipCodeException("Address zip code does not match debit card zip code.");
        }

        if (!validator.hasSufficientFunds(totalAmount)) {
            throw new InsufficientFundsException("Insufficient funds for this purchase.");
        }

        paymentStatus = true;
    }

    // Nested class
    public static class DebitCardValidator {
        private LoginInformation loginInfo;
        private String addressZip;
        private double availableBalance;

        public DebitCardValidator(LoginInformation loginInfo, String addressZip, double availableBalance) {
            this.loginInfo = loginInfo;
            this.addressZip = addressZip;
            this.availableBalance = availableBalance;
        }

        public boolean validateZipCode() {
            return loginInfo.getDebitCardZipCode().equals(addressZip);
        }

        public boolean hasSufficientFunds(double amount) {
            return availableBalance >= amount;
        }
    }
}
```

```python
%%writefile SaleComputation.java
public class SaleComputation {
    private Cart cart;
    private Tax tax;

    public SaleComputation(Cart cart) {
        this.cart = cart;
        this.tax = new Tax();
    }

    public double computeSubtotal() {
        double subtotal = 0;
        for (var item : cart.getItems()) {
            subtotal += item.product.getPrice() * item.quantity;
        }
        return subtotal;
    }

    public boolean verifyStockBeforeCheckout() {
        for (var item : cart.getItems()) {
            if (!item.product.checkStock(item.quantity)) {
                System.out.println("Not enough stock for " + item.product.getProductName());
                return false;
            }
        }
        return true;
    }

    public double computeTotalWithTax() {
        double subtotal = computeSubtotal();
        return tax.totalWithTax(subtotal);
    }
}
```

```python
%%writefile Checkout.java
import java.util.Scanner;

public class Checkout {

    public void processCheckout(Cart cart, LoginInformation loginInfo, double availableBalance) {
        Scanner scanner = new Scanner(System.in);
        SaleComputation sale = new SaleComputation(cart);

        if (!sale.verifyStockBeforeCheckout()) {
            System.out.println("Checkout failed due to stock issues.");
            return;
        }

        double total = sale.computeTotalWithTax();
        System.out.printf("Your total (with tax) is: $%.2f%n", total);

        System.out.print("Do you want to proceed? (yes/no): ");
        String choice = scanner.nextLine();
        if (!choice.equalsIgnoreCase("yes")) {
            System.out.println("Checkout cancelled.");
            return;
        }

        System.out.print("Enter shipping address: ");
        String address = scanner.nextLine();
        System.out.print("Enter shipping zip code: ");
        String addressZip = scanner.nextLine();

        Payment payment = new Payment(total);
        try {
            payment.acceptPayment(loginInfo, addressZip, availableBalance);
            if (payment.isPaymentSuccessful()) {
                // Update stock
                for (var item : cart.getItems()) {
                    item.product.updateStock(item.quantity);
                }
                System.out.println("Payment successful. Order placed!");
            }
        } catch (InsufficientFundsException | InvalidZipCodeException e) {
            System.out.println("Checkout error: " + e.getMessage());
        }
    }
}
```
