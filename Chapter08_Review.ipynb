### 1. Introduction to Arrays:
- An **array** is a collection of elements, all of the same type, stored in contiguous memory locations.
- Each element can be accessed using an **index** (or subscript), which starts at **0**.
- Arrays are useful for storing multiple values when you know how many elements you need.

### 2. Declaring and Initializing Arrays:
- **Declaration**: Similar to declaring any other variable, but include square brackets.


### Initialization:
- Arrays must be explicitly initialized to reserve memory

Arrays can also be initialized with default values:
 - `int` arrays are initialized to 0.
 - `char` arrays are initialized to '\u0000', which is the *null character*.
   The null character ('\u0000') is not the same as `null`. It is a valid character,
   but it does not display anything when printed. This is different from an empty string `""`
   and an array of 0 elements `new int[0]`, both of which still represent valid objects.
   The null character means no data value, while null in Strings mean "no object" or "no reference."
 - `boolean` arrays are initialized to *false*.
 - `String` arrays are initialized to *null*, which means there is no reference to any object.
   When an element in a `String` array is `null`, it indicates that the array slot does not point
   to any actual `String` object, unlike an empty string `""`, which is still a valid `String` object.

````python
%%writefile ArrayExample.java
import java.util.Arrays;

public class ArrayExample {
    public static void main(String[] args) {

        int[] myArray; // declaring an array
        myArray = new int[]{1, 2, 3, 4}; // initialize an array with values
        System.out.println("My first array: " + Arrays.toString(myArray));

        double[] scores = new double[]{80, 82, 85, 90.5}; // another way to initialize Arrays
        System.out.println("My score this semester: " + Arrays.toString(scores));

        int myOtherArray[] = new int[0];
        System.out.println("My other array: " + Arrays.toString(myOtherArray));

        char[] charArray = new char[5];
        System.out.println("My character array: " + Arrays.toString(charArray));

        String[] strArray = new String[5];
        boolean[] boolArr = new boolean[5];

        System.out.println("String array: " + Arrays.toString(strArray));
        System.out.println("Boolean array: " + Arrays.toString(boolArr));

        int[] myArray2 = {50, 16, 20, 15};

        // find the sum of myArray + myArray2
        int[] sumOfArray = new int[myArray.length];

        for(int i = 0; i < myArray.length; i++){
          System.out.println(myArray[i] + ", " + myArray2[i]);
          sumOfArray[i] = myArray[i] + myArray2[i];
        }
        System.out.println("The two arrays when added give us:: " + Arrays.toString(sumOfArray));

    }
}
````
