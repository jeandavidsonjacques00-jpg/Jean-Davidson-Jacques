### 1. Introduction to Arrays:
- An **array** is a collection of elements, all of the same type, stored in contiguous memory locations.
- Each element can be accessed using an **index** (or subscript), which starts at **0**.
- Arrays are useful for storing multiple values when you know how many elements you need.

### 2. Declaring and Initializing Arrays:
- **Declaration**: Similar to declaring any other variable, but include square brackets.


### Initialization:
- Arrays must be explicitly initialized to reserve memory

Arrays can also be initialized with default values:
 - `int` arrays are initialized to 0.
 - `char` arrays are initialized to '\u0000', which is the *null character*.
   The null character ('\u0000') is not the same as `null`. It is a valid character,
   but it does not display anything when printed. This is different from an empty string `""`
   and an array of 0 elements `new int[0]`, both of which still represent valid objects.
   The null character means no data value, while null in Strings mean "no object" or "no reference."
 - `boolean` arrays are initialized to *false*.
 - `String` arrays are initialized to *null*, which means there is no reference to any object.
   When an element in a `String` array is `null`, it indicates that the array slot does not point
   to any actual `String` object, unlike an empty string `""`, which is still a valid `String` object.

````python
%%writefile ArrayExample.java
import java.util.Arrays;

public class ArrayExample{
  public static void main(String[] args){
    int[] myArray; // declaring an array
    myArray = new int[]{1, 2, 3, 4}; // initialize an array with values
    System.out.println("My first array: " + myArray); // values not printing, just memory reference

    double[] scores = {80, 82, 85, 90.5}; // another way to initialize arrays
    System.out.println("My scores this semester: " + Arrays.toString(scores));

    int myOtherArray[] = new int[4];
    System.out.println("My other array: " + Arrays.toString(myOtherArray));

    char[] charArray = new char[5];
    System.out.println("My character array: " + Arrays.toString(charArray));

    String[] strArray = new String[5];
    boolean[] boolArr = new boolean[5];

    System.out.println("String array: " + Arrays.toString(strArray));
    System.out.println("Boolean array: "+ Arrays.toString(boolArr));

    int[] myArray2 = {50, 16, 20, 15};

    // Find the sum of myArray + myArray2
    int[] sumOfArrays = new int[myArray.length];

    for(int i = 0; i < myArray.length; i++){
      System.out.println(myArray[i] + ", " + myArray2[i]);
      sumOfArrays[i] = myArray[i] + myArray2[i];
    }

    System.out.println("The two arrays when added give us: " + Arrays.toString(sumOfArrays));

  }
}
````

````python
%%writefile ArrayAccessExample.java
import java.util.Arrays;

public class ArrayAccessExample{
  public static void main(String[] args){
    int[] myNumbers = {5, 10, 15, 20, 25};

    System.out.println("Element at index 0: " + myNumbers[0]); // output of: 5

    // we can change the values by locating the element location and call its array:
    // Then assing it a new value:
    myNumbers[2] = 17; // Changing the third element to 17
    System.out.println("Modified element at index 2: " + myNumbers[2]);
    for(int i = 0; i < myNumbers.length; i++){
      System.out.println();
      System.out.println("Length of myNumbers array: " + myNumbers.length);

      System.out.println("Element at index " + i + " is: " + myNumbers[i]);
    }

    String[] fruits = {"Apple", "Banana", "Cherry", "Date"};

    System.out.println("nList of fruits:");
    // for each loop:
    for (String fruit: fruits){
      System.out.println(fruit);
    }

    for (String fruit: fruits){
      fruit = "modified";
      System.out.println(fruit);
  }
  System.out.println(Arrays.toString(fruits));

  Person[] people = new Person[3];

  people[0] = new Person("Alice", 23);
  people[1] = new Person("bob", 30);
  people[2] = new Person("Charlie", 22);

  System.out.println("\nList of People:");
  for(Person person: people){
    System.out.println(person.getName() + " is " + person.getAge() + " years old!");
  }
  }
}
class Person{
  // private fields:
  private String name;
  private int age;

  // constructor
  public Person(String name, int age){
    this.name = name;
    this.age = age;
  }

  // creating getters and setters:
  public String getName() {
    return name;
  }
  public void setName(String name){
    this.name = name;
  }

  public int getAge(){
    return age;
  }
  public void setAge(int age){
    this.age = age;
  }
}
````

### 7. Parallel Arrays
- Parallel arrays are multiple arrays where corresponding elements are related. The arrays are not linked, and they do not link elements across arrays. Instead, they simply store related data at the same indices, such as:

```java
String[] names = {"Alice", "Bob", "Charlie"};
int[] scores = {85, 90, 95};
// names[0] corresponds to scores[0]
```

### 8. Passing Arrays to Methods
- Arrays are passed **by reference**, meaning changes inside the method affect the original array. Essentially, it's working with the actual reference of the array, not copies when passing an Array to a method.

```java
public void updateScores(int[] scores) {
    scores[0] = 100; // Changes the original array
}
```

### 9. Returning Arrays from Methods
- Methods can return arrays, allowing you to dynamically create and manipulate arrays within a method.
```java
public int[] generateArray(int size) {
    // 'int[]' indicates that the method returns an array of integers.
    // The parameter 'int size' specifies that the method takes an integer input.
    // This integer 'size' will be used to create a new integer array of the specified size.
    return new int[size];
}
```

### 10. Best Practices and Pitfalls
- **DO NOT** access indices outside the valid range (`0` to `array.length-1`).
- **USE** array.length to avoid hardcoding sizes.
- Be mindful of **DEFAULT VALUES** and **NULL REFERENCES**.
- Avoid assuming that an array of characters (`char[]`) is the same as a String. While both contain characters, a `char[]` and a `String` are distinct in Java, with `String` offering additional methods and features.
  - Even though fundamentally, Strings are multiple characters, `String` is a built in class able to manipulate multiple characters more efficiently and mutliple methods that can concatenate the characters together, look for substrings, search for characters based on their index, and methods compare strings.

````python
%%writefile Solution.java
import java.util.Arrays;
import java.util.HashMap;
import java.util.Map;

public class Solution {
    public static void main(String[] args) {
        int[] nums = {2, 4, 7, 11, 15};
        int target = 9;

        int[] indices = twoSum(nums, target);

        System.out.println("The indices are: " + Arrays.toString(indices));
    }

    public static int[] twoSum(int[] nums, int target) {
        Map<Integer, Integer> map = new HashMap<>();

        for (int i = 0; i < nums.length; i++) {
            int complement = target - nums[i];

            if (map.containsKey(complement)) {
                return new int[] { map.get(complement), i };
            }

            map.put(nums[i], i);
        }

        return new int[] {0, 0};
    }
}
````
