## Understanding Method Calls and Placement

**Method**
- A program module
- Contains a series of statements
- Carries out a task

**Execute a method**
- Invoke or call from another method

**calling Method (client method)**
- Makes a method call

**Called method**
- Invoked by a calling Method

**Main() method executes automatically**
- Other methods are called as needed
- Main() method is also called first

A Method must include:

**Method Header**
- Also called a declaration

**Method body**
- Between a pair of curly braces
- Contains the statements that carry out the work
- Also called implementation

**The method header contains:**
- Optional access specifiers
- Static/non-static
- A return type
- An identifier
- Parentheses
  - might contain data to be sent to the method

**Place the entire method within the class that will use it**
- Not within any other method or outside of any class.

You can place additional methods before and after the `main{}` method

```python
%%writefile First.java
import java.util.Random;

public class First{
  //

  public static void main(String[] args){
    // leaving blank for now
    displayAddress();

    First myFirst = new First(); // creating an instance of the First class
    String myResults = myFirst.getCompanyName(); // to work with the getCompanyName() method
    //String myResults = getCompanyName();
    System.out.println("Your company name is: " + myResults);

    Random randomNumberGen = new Random(); // creating an instance of the random class

    int numA = randomNumberGen.nextInt(11); // 11 is limit, meaning: 0 - 10

    int numB = randomNumberGen.nextInt(11);

    int numC = randomNumberGen.nextInt(11);

    System.out.println("Number 1: " + numA);
    System.out.println("Number 2: " + numB);
    System.out.println("Number 3: " + numC);

    int sumOfNumbers = getMySum(numA, numB, numC); //invoking or calling the method

    System.out.println("The sum of these numbers is: " + sumOfNumbers);
  }

  // Creating 2nd method
  public static void displayAddress(){
    System.out.println("XYZ Company");
    System.out.println("8900 U.S. Hwy 14");
    System.out.println("Crystal Lake, IL 60014");
  }

  public String getCompanyName(){
    return "XYZ Company";
  }

  public static int getMySum(int a, int b, int c){
    int sumOfNums = a + b + c;
    return sumOfNums;
  }
}
```

## Understanding Classes, Objects, and Access Specifiers

**Class**
- A blueprint for creating objects (instances)
- Defines the properties (fields) and behaviors (methods) of the objects it represents

**Object**
- An instance of a class
- Created using the `new` keyword
- Holds its own state (values of its fields) and can perform actions through methods

**Access Specifiers**
- Control the visibility of class members (fields and methods)
- There are four common access specifiers in Java:
  - **Public**: Accessible from any class.
  - **Private**: Only accessible within the same class.
  - **Protected**: Accessible within the same package or by subclasses.
  - **Package-Private (Default)**: Accessible within the same package, but not from outside.

**Encapsulation**
- A principle of restricting direct access to an object's fields
- Achieved using private fields and public getter/setter methods

**Constructor**
- A special method used to initialize objects when they are created
- The constructor's name must match the class name and it doesn't have a return type

### Example: The `Employee` Class

In the following code, we will demonstrate:
- **Fields**: Defining the properties of an employee, such as their name, salary, and position
- **Access Specifiers**: Using `private`, `protected`, and package-private fields
- **Encapsulation**: Restricting direct access to fields and providing controlled access through getter and setter methods
- **Constructors**: Initializing objects with values upon creation
- **Instance Methods**: Methods that belong to objects and work with the object's fields
- **Method Overloading**: Defining multiple versions of a method with different parameters
- **Object Instantiation**: Creating multiple objects with unique values for their fields


```python
%%writefile Employee.java

public class Employee{
  // attributes
  private int empNum;
  private String empFirstName;
  private String empLastName;
  private double empSalary;

  public int getEmpNum(){
    return empNum;
  }

  public void setEmpNum(int empNumber){
    empNum = empNumber;
  }

  public String getEmpFirstName(){
    return empFirstName;
  }

  public void setEmpFirstName(String empFirstName){
    this.empFirstName = empFirstName;
  }

  public String getEmpLastName(){
    return empLastName;
  }

  public void setEmpLastName(String empLastName){
    this.empLastName = empLastName;
  }

  public double getEmpSalary(){
    return empSalary;
  }

  public void setEmpSalary(double empSalary){
    this.empSalary = empSalary;
  }

  public Employee(int empNum, String empFirstName, String empLastName, double empSalary){
    this.empNum = empNum;
    this.empFirstName = empFirstName;
    this.empLastName = empLastName;
    this.empSalary = empSalary;
  }

  // creating a void method to print all employee details:
  public void printEmployeeDetails(){
    System.out.println("Employee number: " + getEmpNum());
    System.out.println("Employee Full Name: " + getEmpFirstName() + " " + getEmpLastName());
    System.out.println("Employee Salary: $" + getEmpSalary());
  }
}
```

## Creating a new class to run the Employee Class

```python
%%writefile DeclareTwoEmployees.java

public class DeclareTwoEmployees{
  public static void main(String[] args){
    Employee clerk = new Employee(0, "James", "James", 10000);
     Employee driver = new Employee(1, "Jones", "Jones", 12000);

     clerk.setEmpNum(123);
     driver.setEmpNum(124);

     System.out.println("The clerk's emp number is: " + clerk.getEmpNum());
     System.out.println("The driver's emp number is: " + driver.getEmpNum());

      clerk.setEmpFirstName("John");
      driver.setEmpFirstName("Jane");

      clerk.setEmpLastName("Doe");
      driver.setEmpLastName("Doe");

      System.out.println("The name of the clerk is: " + clerk.getEmpFirstName() + " " + clerk.getEmpLastName());
      System.out.println("The name of the driver is: " + driver.getEmpFirstName() + " " + driver.getEmpLastName());

      clerk.setEmpSalary(47000);
      driver.setEmpSalary(57000);

      System.out.println("Clerk salary: $" + clerk.getEmpSalary());
      System.out.println("Driver salary: $" + driver.getEmpSalary());
  }
}
```
