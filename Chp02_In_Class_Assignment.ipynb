## Now we're going to go through the Scanner class to accept keyboard input

**System.in** objects:
- They're your gateway to the standard input device
- Usually, that means use of keyboard
- We access this using the `Scanner` class

**Scanner** object
- This basically breaks input into units called `tokens`.

So, what we'll do is basically write some code to use the `System.in` to be able to accept user input, and we'll use the `Scanner` object to basically prompt the user with an interface to be able to feed input.

### Activity

**Task: Create a Simple User Input Program**

1. **Write a Java Program:**
  - Write a program that asks the user for their name, age and favorite color.

2. **User Scanner for user Input:**
  - Use the `Scanner` class to get these inputs from the user.

3. **Print the Input:**
  - After getting the inputs from the user, print them out in a formmated message.


```python
## Code to get you started
%%writefile UserInputDemo.java
import java.util.Scanner;

public class UserInputDemo{
  public static void main(String[] args){
    // Creating a 'Scanner' class object:
    // Rules when creating an object:
    // class_name object_name_you_come_up_with = new_memory_space class_constructor
    Scanner input = new Scanner(System.in);

    // the object is 'input'
    System.out.print("Enter your name: ");

    String name = input.nextLine();

    // Ask user for age and favorite color:
    System.out.print("Enter your age: ");

    int age = input.nextInt();
    input.nextLine(); // This is used to consume the input leftover '\n'

    System.out.print("Enter your favorite color: ");

    String color = input.nextLine();

    // printing the collected data from user:
    System.out.println("Hello, " + name + "!");
    System.out.println("You are: " + age + " years old.");
    System.out.println("Your favorite color is: " + color + ".");

    // close the scanner object 'input':
    input.close();
  }
}
```
